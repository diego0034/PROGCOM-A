```python
import math

print("SELECCIONA EL TIPO DE OPERACIÓN CON VECTORES")
print("")
print("1- Magnitud de un vector")
print(" OPERACIONES ENTRE VECTORES")
print("2- Suma y Resta")
print("3- Producto Punto")
print("4- Producto Cruz")
print("5- Multiplicación por un escalar")
print("_________________________________")

x=int(input("Eliga ----> "))

if x==1:
  print("___Seleccionaste Magnitud de vector___")
  opc=str(input("Es en R2 o R3? ")).lower()
  if opc=="r2":
   v1=int(input("Escribe el valor de x del vector-> "))
   v2=int(input("Escribe el valor de y del vector-> "))
   m=math.sqrt((v1**2)+(v2**2))
   print("La magnitud del vector es", m)
  elif opc=="r3":
   v1=int(input("Escribe el valor de x del vector-> "))
   v2=int(input("Escribe el valor de y del vector-> "))
   v3=int(input("Escribe el valor de z del vector-> "))
   m=math.sqrt((v1**2)+(v2**2)+(v3**2))
   print("La magnitud del vector es", m)
  else:
   print("Plano no válido")
   
elif x==2:
  print("___Suma y resta___")
  opc=str(input("Es en R2 o R3? ")).lower()
  if opc=="r2":
   v11=int(input("Escribe el valor de x del vector 1-> "))
   v12=int(input("Escribe el valor de y del vector 1-> "))
   print("_____________________________________________")
   v21=int(input("Escribe el valor de x del vector 2-> "))
   v22=int(input("Escribe el valor de y del vector 2-> "))
   sumax=(v11+v21)
   sumay=(v12+v22)
   print("Los valores del nuevo vector son en X",sumax,"y en Y", sumay)
  elif opc=="r3":
   v11=int(input("Escribe el valor de x del vector 1-> "))
   v12=int(input("Escribe el valor de y del vector 1-> "))
   v13=int(input("Escribe el valor de z del vector 1-> "))
   print("_____________________________________________")
   v21=int(input("Escribe el valor de x del vector 2-> "))
   v22=int(input("Escribe el valor de y del vector 2-> "))
   v23=int(input("Escribe el valor de z del vector 2-> "))
   sumax=(v11+v21)
   sumay=(v12+v22)
   sumaz=(v13+v23)
   print("Los valores del nuevo vector son en X ",sumax,"en Y",sumay,"y en Z", sumaz)
  else:
   print("Plano no válido")
elif x==3:
  print("___Seleccionaste Producto punto___")
  print("VECTOR 1")
  v1=int(input("Escribe el valor de x del vector 1-> "))
  v2=int(input("Escribe el valor de y del vector 1-> "))
  v3=int(input("Escribe el valor de z del vector 1-> "))
  print("VECTOR 2")
  z1=int(input("Escribe el valor de x del vector 2-> "))
  z2=int(input("Escribe el valor de y del vector 2-> "))
  z3=int(input("Escribe el valor de z del vector 2-> "))
  producto_punto=(v1*z1)+(v2*z2)+(v3*z3)
  print("El producto punto de esos 2 vectores es", producto_punto)

elif x==4:
  print("___Seleccionaste Producto cruz___")
  print("VECTOR 1")
  v1=int(input("Escribe el valor de x del vector 1-> "))
  v2=int(input("Escribe el valor de y del vector 1-> "))
  v3=int(input("Escribe el valor de z del vector 1-> "))
  print("VECTOR 2")
  z1=int(input("Escribe el valor de x del vector 2-> "))
  z2=int(input("Escribe el valor de y del vector 2-> "))
  z3=int(input("Escribe el valor de z del vector 2-> "))
  nx= (v2*z3)-(v3*z2)
  ny=-(v1*z3)-(v3*z1)
  nz= (v1*z2)-(v2*z1)
  print(nx,"en x",ny,"en y",nz,"en z son los valores del nuevo vector")
elif x==5:
 print("___Seleccionaste Multiplicación por un escalar___")
 opc=str(input("Es en R2 o R3? ")).lower()
 if opc=="r2":
   v1=int(input("Escribe el valor de x del vector-> "))
   v2=int(input("Escribe el valor de y del vector-> "))
   e=int(input("Cual es el escalar-> "))
   nv1=v1*e
   nv2=v2*e
   print("El nuevo vector es", nv1 ,"en X",nv2, "en Y" )
 elif opc=="r3":
   v1=int(input("Escribe el valor de x del vector-> "))
   v2=int(input("Escribe el valor de y del vector-> "))
   v3=int(input("Escribe el valor de z del vector-> "))
   e=int(input("Cual es el escalar-> "))
   nv1=v1*e
   nv2=v2*e
   nv3=v3*e
   print("El nuevo vector es", nv1 ,"en X", nv2 , "en Y y ", nv3 ,"en Z")
 else:
   print("Plano no válido")

else:
  print("___Opcion no válida___")

```
